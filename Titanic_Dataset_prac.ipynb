Treating Null Values


```python
# Filling the Null Values in column "Age" with the mean values
df["Age"].fillna(df['Age'].mean(),inplace=True)
```

```python
# Lets look at the Null Values again
df['Age'].isnull().sum()
```

```python
# Dropping the Column "Cabin" from the dataset since we wont be using it at all.
df.drop('Cabin',axis=1,inplace=True)
```

Converting Categorical Data to Numerical Data

```python
# Checkin the Dtypes of all the columns in the Dataset
df.dtypes
```

As we can see the column's "Name","Sex","Ticket" and "Embarked" are of object datatypes. One way to take care of this problem is by using One-Hot Encoding to change the categorical data into numerical data.

```python
gender = pd.get_dummies(df['Sex'],drop_first=True)
```

```python
df['Gender']=gender
```

```python
df.head()
```

```python
# Now we can finally drop the data that is not usefull in the prediction
df.drop(['Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# Splitting the Independent Variables and Dependent Variables
x=df[['PassengerId','Pclass','Age','SibSp','Parch','Fare','Gender']]
y=df['Survived']
```

```python
y.value_counts()
```

```python
# Importing the train_test_split and Splitting the data into test and train
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.25,random_state=100)
```

```python
# Importing the Model Algo i.e. Logistic Regression which is mainly used for binary classification
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
model_pred_train = lr.predict(x_train)
```

```python
model_pred_test = lr.predict(x_test)
```

```python
# Deploying the Confusion Matrix
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test,model_pred_test),columns=['Predicted No','Predicted Yes'],index=['Actual No','Actual Yes'])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,model_pred_test))
```
